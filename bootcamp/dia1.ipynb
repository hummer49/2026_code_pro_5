## Challenge #2

* Suma: 
    * Asignar dos números en dos variables diferentes `cantidad_de_manzanas` y `cantidad_de_naranjas`. 
    * Luego sumar ambas variables en `cantidad_de_frutas`.
    * Imprimir la última variable.

* Resta: 
    * Asignar un valor numérico a la variable `frutas_vendidas`. 
    * Luego, asignar la resta entre `cantidad_de_frutas` y `frutas_vendidas` a la variable `frutas_restantes`. 
    * Imprimir la última variable.

```python
cantidad_de_manzanas = 10
cantidad_de_naranjas = 15

cantidad_de_frutas = cantidad_de_manzanas + cantidad_de_naranjas
print(f"Cantidad de frutas: {cantidad_de_frutas}")


frutas_vendidas = 20

frutas_restantes = cantidad_de_frutas - frutas_vendidas
print(f"Frutas restante : {frutas_restantes}")
```

## Challenge #6
Escribir una funcion que nos pregunte las caracteristicas de un elemento a comprar e imprima el elemento con sus caracteristicas. 

Debes usar parametros y argumentos 

```python
# definimos la funcion
def caracteristicas_elemento(nombre:str)->None:
    """
    Input:
        nombre (str) -- nombre del elemento a describir
    Output:
        None
    """
    caracteristicas = input(f"Describa las caracteristicas del elemento >>{nombre}<<: ")
    print(f"El elemento {nombre} tiene las siguientes caracteristicas:\n{caracteristicas}")

# llamamos la funcion
nombre_elemento = "pelota"
caracteristicas_elemento(nombre = nombre_elemento)
```

## Challenge #7
Objetivo:
* Escribir una función que pregunte por las características de un producto a comprar (por ejemplo, tipo, tamaño, color) e imprima un resumen de ese producto con las características ingresadas.

Además, deben trabajar con variables locales y globales.

```python
# Definimos una variable global (en ese caso el productoa a describir)
PRODUCTO = "pelota"

# definimos la funcion
def describir_producto():
    tipo = input("Ingrese el tipo del producto {}: ".format(PRODUCTO))
    tamanho = input("Ingrese el tamanho del producto {}: ".format(PRODUCTO))
    color = input("Ingrese el color del producto {}: ".format(PRODUCTO))
    print("Las caracteristicas de {} son:\n- Tipo: {}\n- Tamanho: {}\n- Color: {}".format(PRODUCTO, tipo,tamanho,color))

describir_producto()
```
