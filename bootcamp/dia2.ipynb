## Challenge #1
Escribe un programa que pregunte al usuario por su edad y determine si tiene edad suficiente para comprar y consumir alcohol.
* Si el usuario tiene mas de 18 pero menos de 21 impirmir `"Podes tomar pero no comprar"`
* Si el usuario tiene menos de 18 imprimir `"Juguito nomas"`
* Si el usuario tiene mas de 21 imprimir `"Hace lo que quieras"`

```python
# Preguntar al usuario su edad
edad = int(input("Por favor ingresar la edad del usuario: ")) # asumimos que es un entero
print("La edad del usuario es: {}".format(edad))
# Evaluamos las condiciones
if edad < 18:
    print("Juguito nomas")
elif edad > 21:
    print("Hagan lo que quieran")
else:
    print("Puede tomar pero no comprar")
```

## Challenge #2
Escribir un programa en el que se pregunte al usuario por una temperatura (en grados Celsius), imprima la temperatura escrita, y luego clasificar la temperatura ingresada en los siguientes niveles:

* Entre `-10 y 18` grados, que imprima `"Fresquete"`.
* Entre `19 y 30` grados, que imprima `"Calor'i"`.
* Entre `31 y 45` grados, que imprima `"Hakuuuuuu"`.
* Para todo lo demás, imprimir `"Ñamano mba'e!"`.

```python
# Definimos la funcion que clasifica la temperatura
def clasificar_temperatura(temperatura):
    if temperatura >= -10 and temperatura <= 18:
        print("Fresquete")
    elif temperatura >= 19 and temperatura <= 30:
        print("Calor'i")
    elif temperatura >= 31 and temperatura <= 45:
        print("Hakuuuuuu")
    else:
        print("Ñamano mba'e")

# Pedimos al usuario ingresar la temperatura
temperatura = int(input("Por favor ingrese la temperatura (como numero entero):"))
# mostramos la temperatura ingresada
print("La temperatura ingresada es: {}".format(temperatura))
# Clasificamos la temperatura
clasificar_temperatura(temperatura=temperatura)
```
