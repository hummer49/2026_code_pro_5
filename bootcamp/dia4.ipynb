### Challenge #1
Escribir un programa que cree un `diccionario vacío` y lo vaya llenado con información sobre una persona (por ejemplo nombre, edad, sexo, teléfono, correo electrónico, etc.) que se le `pida al usuario`. 

Cada vez que se `añada` un nuevo dato debe `imprimirse` el contenido del diccionario.

```python
# Creamos el diccionario vacio
persona = {}

# Agregar nombre
nombre_input = input("Ingrese un nombre: ")
persona["nombre"] = nombre_input
print(persona)

# Agregar apellido
apellido_input= input("Ingrese un apellido: ")
persona["apellido"] = apellido_input
print(persona)

# Agregar ciudad de residencia
ciudad_input= input("Ingrese una ciudad: ")
persona["ciudad"] = ciudad_input
print(persona)

# Agregar email
email_input= input("Ingrese un correo: ")
persona["email"] = email_input
print(persona)
```

```python
# Creamos un diccionario vacio
persona_1 = {}

# Definimos los campos que va a ingresar el usuario
campos = ["nombre", "apellido", "ciudad", "email"]

# Iteramos sobre cada campo
for campo in campos:
    # Pedimos al usuario que ingrese el campo actual
    val = input(f"Ingrese {campo}: ")
    # Incluimos el par clave:valor al diccionario
    persona_1[campo] = val
    # Mostramos el estado actual del diccionario
    print(persona_1)
```

### Challenge #2

Crear una tarjeta personal. El usuario (mediante `input`) puede elegir:

1. `Crear` datos nuevos
2. `Modificar` datos
3. `Eliminar` datos

```python
tarjeta_personal = {}
while True:
    comando = input("Presione a (crear), b (modificar), c(eliminar), q(salir)").strip()
    # Crear datos
    if comando == 'a':
        print("\nCrear dato")
        # ingresar clave
        clave = input("Ingrese la nueva clave:")
        # validar si clace existe
        if clave in tarjeta_personal.keys():
            print(f"Error. La clave < {clave} > ya existe")
            continue
        else:
            valor = input("Ingrese el valor: ")
            tarjeta_personal[clave] = valor
    # =======================================================
    # Modificar datos
    elif comando == 'b':
        print("\nModificar dato")
        print(f"Las claves disponibles son: {[x for x in tarjeta_personal.keys()]}")
        clave = input("Ingrese la clave del valor a modificar: ")
        if clave not in tarjeta_personal.keys():
            print(f"Error. La clave < {clave} > no existe")
            continue
        else:
            valor = input(f"Ingrese el nuevo valor para la dato < {clave} >")
            tarjeta_personal[clave] = valor
    # =======================================================
    # Eliminar datos
    elif comando == 'c':
        print("\nEliminar dato")
        print(f"Las claves disponibles son: {[x for x in tarjeta_personal.keys()]}")
        clave = input("Ingrese la clave del valor a eliminar: ")
        if clave not in tarjeta_personal.keys():
            print(f"Error. La clave < {clave} > no existe")
            continue
        else:
            print(f"Eliminando dato < {clave} >")
            del tarjeta_personal[clave]
    # =======================================================
    # Salir del loop
    elif comando == 'q':
        print("Saliendo del programa")
        break
    # =======================================================
    # Volver a comenzar el ciclo
    else:
        print(f"Error. El comando < {comando} > no es valido. Vuelva a ingresar un comandao")
        continue
    # Mostrar el estado del diccionario al fin de cada loop
    print(f"El diccionario es:\n{tarjeta_personal}")
```

### Challenge #3
Crear una agenda de contactos. El usuario puede elegir:

1. `Mostrar` los contactos
2. `Crear `contacto
3. `Modificar` datos del contacto

```python
contactos = {}
while True:
    comando = input("Presione a (crear), b (modificar), c(mostrar), q(salir)").strip()
    # Crear contactos
    if comando == 'a':
        print("\nCrear contacto")
        # ingresar clave
        clave = input("Ingrese la nueva clave:")
        # validar si clace existe
        if clave in contactos.keys():
            print(f"Error. La clave < {clave} > ya existe")
            continue
        else:
            valor = input("Ingrese el valor: ")
            contactos[clave] = valor
    # =======================================================
    # Modificar contactos
    elif comando == 'b':
        print("\nModificar contacto")
        print(f"Las claves disponibles son: {[x for x in contactos.keys()]}")
        clave = input("Ingrese la clave del valor a modificar: ")
        if clave not in contactos.keys():
            print(f"Error. La clave < {clave} > no existe")
            continue
        else:
            valor = input(f"Ingrese el nuevo valor para el contacto < {clave} >")
            contactos[clave] = valor
    # =======================================================
    # Mostrar contactos
    elif comando == 'c':
        print("\nMostrar contactos")
        for key, value in contactos.items():
            print(f"{key}: {value}")
    # =======================================================
    # Salir del loop
    elif comando == 'q':
        print("Saliendo del programa")
        break
    # =======================================================
    # Volver a comenzar el ciclo
    else:
        print(f"Error. El comando < {comando} > no es valido. Vuelva a ingresar un comandao")
        continue
    
```

Luego, crear nuevos objetos a partir de las nuevas class. Imprimir el nombre y el color de los animales en una frase.
Ejemplo: "Me llamo `{nombre}` y soy un ave de color `{color}`".

```python
# definimos la clase
class Animal:
    reino = 'Animalia' 
    # definimos el constructor
    def __init__(
        self,
        nombre,
        color,
        especie
    ):
        self.nombre = nombre
        self.color = color
        self.especie = especie

# Instanciamos las clases
animal_1 = Animal("Tommy", "blanco", "gato")
print(f"Me llamo {animal_1.nombre} y soy un {animal_1.especie} de color {animal_1.color}")
animal_2 = Animal("Michi", "gris", "perro")
print(f"Me llamo {animal_2.nombre} y soy un {animal_2.especie} de color {animal_2.color}")
```

### Challenge #5
Crear dos métodos para nuestro animal y ejecutar, debe ser algo específico de ese animal.

Ejemplo: "Cuando hablo, digo X".

Ejemplo: "Cuando me muevo, hago X".

```python

class Mascota:
    def __init__(
        self,
        nombre,
        especie,
        sonido = None,
        movimiento = None
    ):
        self.nombre = nombre
        self.especie = especie
        self.sonido = sonido
        self.movimiento = movimiento
    
    def hablar(self):
        print(f"Cuando hablo, digo {self.sonido}")
    def moverse(self):
        print(f"Cuando me muevo, {self.movimiento}")


mascota_1 = Mascota("Michi", "gato", "miau", "muevo la cola")

mascota_1.hablar()
mascota_1.moverse()
```

### Challenge #6
Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

Observación: Editar las clases hijas para que no se inicialicen.

```python

class Planta:
    def __init__(
        self,
        especie,
        caracteristicas
    ):
        self.especie = especie
        self.caracteristicas = caracteristicas

class Oleaginosa(Planta):
    clasificacion = "Oleaginosa"
    def mostrar_caracteristicas(self):
        print(f"La planta de {self.especie} es una {self.clasificacion}, y se caracteriza por {self.caracteristicas}")
    def extraer_aceite(self, cantidad):
        print(f"De la planta de {self.especie} se extraen {cantidad} lts de aceite")


# Instanciamos un objeto
oleaginosas_1 = Oleaginosa("Glycine max (soja)", "ser utilizada en industria, energia, alimentacion")

oleaginosas_1.mostrar_caracteristicas()
oleaginosas_1.extraer_aceite(10)
```

```python
class Persona:
    def __init__(
        self,
        nombre: str | None ,
        apellido: str | None,
        ciudad: str | None = None
    ):
        self._nombre = nombre
        self._apellido = apellido
        self._ciudad = ciudad
    # ===
    def set_nombre(self, nombre):
        self._nombre = nombre
    def get_nombre(self):
        return self._nombre
    # ===
    def set_apellido(self, apellido):
        self._apellido = apellido
    def get_apellido(self):
        return self._apellido
    # ===
    def set_ciudad(self, ciudad):
        self._ciudad = ciudad
    def get_ciudad(self):
        return self._ciudad
    
class Ciudadano(Persona):
    def ir_a_votar(self):
        print(f"{self.get_nombre()} {self.get_apellido()} va a ir a votar a {self.get_ciudad()}")


ciudadano_1 = Ciudadano("Pedro", "Perez", "Luque")

ciudadano_1.ir_a_votar()
```

### Challenge #7
Crear dos personajes de una historia que se representen a traves de clases. 
* Ejemplo: Un hereo y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

* Ejemplo: Yo me llamo pepito y tengo super fuerza
* Ejemplo: Yo me llamo juanito y puedo volar

```python
# Crear una clase madre Personaje, con atributos y metodos en comun
class Personaje:
    tipo = "personaje"
    def __init__(self
        ,nombre
        ,poder
    ):
        self.nombre = nombre  
        self.poder = poder
    def presentarse(self):
        print(f"Su nombre es {self.nombre} y su poder es {self.poder}, es un {self.tipo}")
    
# Crear una clase hija Heroe, que herede de la clase Personaje
class Heroe(Personaje):
    tipo = "heroe"
    def hacer_bien(self):
        print(f"El {self.tipo} llamado {self.nombre} hace el bien")
# Crear una clase hija Villano, que herede de la clase Personaje
class Villano(Personaje):
    tipo = "villano"
    def hacer_mal(self):
        print(f"El {self.tipo} llamado {self.nombre} hace el mal")


heroe_1 = Heroe("Pepito", "volar")
heroe_1.presentarse()


villano_1 = Villano("Mamerto", "correr")
villano_1.presentarse()
```
