### Challenge #2
Crear una función que reciba dos letras (parámetros) y que verifique si son iguales:

1. Si son iguales, imprimir `"Las letras son iguales"`.

2. De lo contrario, imprimir `"Las letras son diferentes"`.

````python
# definimos la funcion
def comparar_letras(l1,l2):
    if l1==l2:
        print("Las letras son iguales")
    else:
        print("Las letras son diferentes")


# creamos las variables
letra_1 = input("Ingrese una letra")
print(f"Letra 1 : {letra_1}")
letra_2 = input("Ingrese una letra")
print(f"Letra 2 : {letra_2}")
# llamamos a la funcion
comparar_letras(letra_1, letra_2)
````

### Challenge #3

Crear una función que reciba dos parámetros (una palabra y un numero), y que retorne `True` si el número es igual a la cantidad de letras de la palabra. En caso contrario, que retorne `False`.

Pista: cantidad de letras de una palabra en Python

````python
# Definimos la funcion
def comprobar_longitud(palabra, numero):
    if len(palabra) == numero:
        return True
    else:
        return False

# probamos la funcion
palabra_prueba = input("Introduce una palabra:")
print(f"La palabra ingresada es: {palabra_prueba}")
numero_prueba = int(input("Introduce un numero entero: "))
print(f"El numero ingresado es: {numero_prueba}")
# Llamamos a la  funcion
evaluacion = comprobar_longitud(palabra_prueba, numero_prueba)
print(f"La longitud de la cadena es igual a {numero_prueba}? {evaluacion}")
# print(f"La longitud de la cadena es igual a {numero_prueba}? {'Si' if evaluacion else 'No'}")
````

### Challenge #4
Crear una función que revise si una palabra contiene una letra. 
* En caso de que la contenga, retornar la letra entre corchetes []. 
* De lo contrario, retornar solo la letra.

````python
# definimos la funcion
def letra_en_palabra(palabra, letra):
    if letra in palabra:
        return f"[{letra}]"
    else:
        return letra
    
# Probamos la funcion
palabra_prueba = input("Introduce una palabra: ")
print(f"La palabra ingresada es: {palabra_prueba}")
letra_prueba = input("Introduzca una letra: ")
print(f"La letra ingresada es: {letra_prueba}")
# Llamamos la funcion
evaluacion = letra_en_palabra(palabra_prueba, letra_prueba)
print(evaluacion)
````

## Challenge #5

* Pedir al usuario que ingrese una palabra, luego convertir esa palabra a una lista con sus caracteres con `list()`. 
* Agregar esa lista de caracteres a una lista llamada `palabras_ingresadas`. 
* Repetir el proceso 3 veces.
* Cada vez que se agregue una lista de caracteres, imprimir la lista `palabras_ingresadas`.

Ejemplo del resultado esperado:

```[['c', 'i', 'e', 'l', 'o']]``` # Al ingresar la primera palabra

```[['c', 'i', 'e', 'l', 'o'], ['c', 'a', 'l', 'o', 'r']]``` # Al ingresar la segunda palabra

```[['c', 'i', 'e', 'l', 'o'], ['c', 'a', 'l', 'o', 'r'], ['n', 'u', 'b', 'e', 's']]``` # Al ingresar la tercera palabra

````python
# inicializamos una lista vacia para guardad lista de palabras
palabras_ingresadas = []
# iniciamos un loop
for ii in range(3):
    # pedimos al usuario ingresar la palabra
    palabra = input("Ingrese una palabra")
    # convertimos la palabra en una lista de caracteres
    palabra_lista = list(palabra)
    # agregamos la palabra_lista a la lista palabras_ingresadas
    palabras_ingresadas.append(palabra_lista)
    # mostramos palabras_ingresadas
    print(palabras_ingresadas)
````

```
['h', 'o', 'l', 'i', 's']
['h', 'e', 'l', 'l', 'o']
['h', 'a', 'l', 'l', 'o']
```

````python
def imprimir_grilla(grilla):
    for ii in grilla:
        print(ii)

lista = [['h', 'o', 'l', 'i', 's'], ['h', 'e', 'l', 'l', 'o'], ['h', 'a', 'l', 'l', 'o']]
imprimir_grilla(lista)
````
